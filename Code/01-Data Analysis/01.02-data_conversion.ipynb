Convert to CSV

```python
with open("filtered_reviews_Cell_Phones_and_Accessories_5.json", 'rb') as f:
    data = f.readlines()
    
data = map(lambda x: x.rstrip(), data)
df = pd.DataFrame()
for r in data:
    r = eval(r)
    s  = pd.Series(r,index=r.keys())
    df = df.append(s,ignore_index=True)
df.to_csv('Cell_Phones_and_Accessories.csv', index=False)
```
