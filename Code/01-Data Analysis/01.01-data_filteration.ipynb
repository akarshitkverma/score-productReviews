Due to limited computational limit, select reviews if the product it is associated with has more than 10 reviews and has more than 5 votes.

```python
productReviewCount={}
count=0

for line in open("reviews_Cell_Phones_and_Accessories_5.json", 'r'):
    review=(json.loads(line))
    product = review["asin"]
    if (product in productReviewCount):
        productReviewCount[product] += 1
    else:
        productReviewCount[product] = 1

filtered = open("filtered_reviews_Cell_Phones_and_Accessories_5.json",'a')

for line in open("reviews_Cell_Phones_and_Accessories_5.json", 'r'):
    review = (json.loads(line))
    product = review["asin"]
    votes = review['helpful'][0]+review['helpful'][1]
    if((product in productReviewCount) and (productReviewCount[product]>10)):
      if (votes>5):
        count+=1
        filtered.write(line)

print("Number of Filtered Reviews are "+str(count))
```
