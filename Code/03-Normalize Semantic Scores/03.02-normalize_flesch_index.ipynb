Since minimum and maximum value of 'flesch_index' are not 0 and 1 respectively, the 'flesch_index' needs to be computed in normalized format. Thus, we will perform following steps of normalization on it.

Add new column 'normalized_Flesch_Index' to dataframe

```python
df['normalized_flesch_index']=0
```

Compute normalized flesch index for each review and store value in column 'normalized_Flesch_Index'

```python
length=len(df)
minIndex=df['flesch_index'].min()
maxIndex=df['flesch_index'].max()
for i in range(length):
  index=df.iloc[i,8]
  normalizedIndex=(index-minIndex)/(maxIndex-minIndex)
  df.iloc[i,14]=normalizedIndex
```
