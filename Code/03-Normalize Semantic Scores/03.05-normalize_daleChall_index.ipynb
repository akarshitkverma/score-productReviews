Since minimum and maximum value of 'daleChall_index' are not 0 and 1 respectively, the 'daleChall_index' needs to be computed in normalized format. Thus, we will perform following steps of normalization on it.

Add new column 'normalized_daleChall_index' to dataframe

```python
df['normalized_daleChall_index'] = 0
```

```python
length=len(df)
minIndex=df['daleChall_index'].min()
maxIndex=df['daleChall_index'].max()
for i in range(length):
  index=df.iloc[i,11]
  normalizedIndex=(index-minIndex)/(maxIndex-minIndex)
  df.iloc[i,17]=normalizedIndex
```
