Since minimum and maximum value of 'entropy_score' are not 0 and 1 respectively, the 'entropy_score' needs to be computed in normalized format. Thus, we will perform following steps of normalization on it.

Add new column 'normalized_entropy_score' to dataframe

```
df['normalized_entropy_score']=0
```

```
length=len(df)
minIndex=df['entropy_score'].min()
maxIndex=df['entropy_score'].max()
for i in range(length):
  index=df.iloc[i,10]
  normalizedIndex=(index-minIndex)/(maxIndex-minIndex)
  df.iloc[i,16]=normalizedIndex
```
