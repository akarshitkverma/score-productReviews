Compute helpful ratio for each review and store value in column 'helpful_ratio'

```python
reviewer_map={}

def calculate_votes(row):
    helpfulstr = row['helpful']
    helpfulstr = helpfulstr.strip('[]')
    helpfullist = helpfulstr.split(',')
    helpfullist = [int(num) for num in helpfullist]
    #print(helpfullist)
    reviewerID = row['reviewerID']
    if reviewerID in reviewer_map.keys():
        reviewer_map[reviewerID][0] += helpfullist[0]
        reviewer_map[reviewerID][1] += helpfullist[1]
    else:
        reviewer_map[reviewerID] = helpfullist
        

for index,row in df.iterrows():
    calculate_votes(row)
```

```python
reviewer_map
```

```python
def set_helpful_ratio(reviewer_id):
    helpfullist = reviewer_map[reviewer_id]
    helpfullist[0] += 1
    helpfullist[1] += 2
    helpfulratio = helpfullist[0]/helpfullist[1]
    return helpfulratio
```

```python
df['helpful_ratio'] = df['reviewerID'].apply(set_helpful_ratio)
```
