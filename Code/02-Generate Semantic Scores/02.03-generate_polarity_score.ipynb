Compute Polarity score for each review and store value in column 'polarity_score'


```python
!pip install textblob
```

```python
from textblob import TextBlob
```

```python
length = len(df)
for i in range(length):
  review = df.iloc[i,6]
  polarityScore = TextBlob(review).sentiment.polarity
  df.iloc[i,9] = polarityScore
```
