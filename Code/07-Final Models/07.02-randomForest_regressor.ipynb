Read dataset obtained from previous step where semantic scores are calculated

```python
df=pd.read_csv("Cell_Phones_and_Accessories.csv")
```

```python
df.head(4)
```

Creating list of parameters/variables

```python
X_columns = ['subjectivity_score', 'normalized_flesch_index', 'normalized_polarity_score', 'normalized_entropy_score', 'normalized_daleChall_index', 'lexDiversity_ratio', 'helpful_ratio']
```

Segregating the columns

```python
X = df[X_columns]
Y = df['utility_score']
```

Splitting data into training and testing dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(X, Y, test_size=0.2, random_state=1)

print("Train data points: ", x_train.shape[0])
print("test data points: ", x_test.shape[0])
```

Building the model

```python
from sklearn.ensemble import (RandomForestRegressor)
from sklearn.metrics import mean_squared_error, r2_score
```

Random Forest Regressor with tuned parameter from previous step

```python
model = RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=4, max_features='auto', max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=100, n_jobs=None, oob_score=False,
                      random_state=1, verbose=0, warm_start=False)
```

Fitting the model to the training set

```python
model.fit(x_train,y_train)
model.score(x_train,y_train)
```

Predicting for training set

```python
y_pred = model.predict(x_train)
RMSE = np.sqrt(mean_squared_error(y_train, y_pred))
R_sq = r2_score(y_train, y_pred)
```

Scores

```python
print("RMSE in testing:", RMSE)
```

```python
print("R sqaured in testing:", R_sq)
```

Adding predicted value column to the dataframe for visualization

```python
df['randomForest_regressor'] = model.predict(X)
```

```python
df.to_csv('Cell_Phones_and_Accessories.csv',index=False)
```
